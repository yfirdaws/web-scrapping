```python
import requests
from requests import get
#from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time 
from random import randint
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
opts = Options()
opts.headless = False
driver = webdriver.Chrome(ChromeDriverManager().install(),options=opts)
```

```python
Name=[]
Price=[] 
url="https://www.supermart.ng/search?q=Food%20cupboard"
driver.get(url)
time.sleep(10)
#click show price button
button = driver.find_element_by_xpath('//*[@id="storeBody"]/app-search/div/div[2]/div[3]/div[1]/app-product-item/div/div/div[2]/div[1]/button')
button.click()
time.sleep(10)
#select first location from list
location=driver.find_element_by_xpath('//*[@id="accordionLocationheading0"]/h2/button')
location.click()
#click continue
cont=driver.find_element_by_xpath('//*[@id="mat-dialog-0"]/app-location-check/div/div/button')
cont.click()
#search food cupboard in search box
time.sleep(10)
search_box = driver.find_element_by_xpath('//*[@id="target"]/div/input')
search_box.send_keys('Food cupboard') 
search_box.send_keys(Keys.ENTER)
page=1
while page<80:
    url="https://www.supermart.ng/search?q=Food%20cupboard&page={}".format(page)
    try:
        driver.get(url)
    except TimeoutException as ex:
        print(ex.Message)
        driver.navigate().refresh()
   # time.sleep(10)
    #ext_page_btn =driver.find_elements_by_link_text('Next')
   #if len(next_page_btn) <1:
   #    print("no more pages left")
   #    break
    #lse:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")#scrolls through page to ensure pge loads before scraping
    WebDriverWait(driver, 20)
    #pWebDriverWait(driver,10).until(EC.presence_of_all_elements_located(By.xpath("//*[@id='storeBody'']/app-search/div/div[2]/div[3]/div[1]/app-product-item/div/div/div"))) #wait allows the program to pause till the page is loaded dynamically
    #products=driver.find_element_by_class_name("card-body.product-body.d-flex.flex-column")
    #for item in products:
    #wait.until(presence_of_all_elements_located(By.className('product-col.mx-1'))
    title=driver.find_elements_by_class_name('p-line-camp')
    for n in title:
        if n is not None:
            name=n.text
        else:
            name=np.nan
        Name.append(name)
    #print(Name)
    #time.sleep(1)
    price=driver.find_elements_by_class_name('col-5.clear-col-lr.product-price')
    for p in price:
        if p is not None:
            amount=p.text
        else:
            amount=np.nan
        Price.append(amount)
    print(Name,Price)
    page=page+1
    #time.sleep() 
  # nxt=DriverWait(driver,5).until(EC.element_to_be_clickable((By.LINK_TEXT, "Next")))
  # driver.execute_script("return arguments[0].scrollIntoView();", nxt)
  # nxt.click()
```
