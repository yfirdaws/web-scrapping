```python
import requests
from requests import get
#from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time 
from random import randint
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
opts = Options()
opts.headless = False
driver = webdriver.Chrome(ChromeDriverManager().install(),options=opts)
```

```python
page=1
Name=[]
Price=[]
while page !=51:
    url=f"https://www.konga.com/category/groceries-581?page={page}"
    try:
        driver.get(url);
    except TimeoutException as ex:
        print(ex.Message)
        driver.navigate().refresh()
    #time.sleep(10)
   # time.sleep(15)
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")#scrolls through page to ensure pge loads before scraping
    WebDriverWait(driver,20)

    products= WebDriverWait(driver,20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div._4941f_1HCZm")))
    for p in products:
        #title=p.find_elements_by_class_name('af885_1iPzH')
        if p is not None:
            name=p.find_element_by_class_name('af885_1iPzH').text
        else:
            name=None
        Name.append(name)
    #print(Name)
    #time.sleep(1)
        #price=p.find_elements_by_class_name('d7c0f_sJAqi')
    #for p in price:
        if p is not None:
                amount=p.find_element_by_class_name('d7c0f_sJAqi').text
        else:
            amount=None
        Price.append(amount)
    print(Name,Price)
   # time.sleep(15) 
        
    page=page+1
```

```python
n=pd.DataFrame({'Name':Name})
it=pd.DataFrame({'Price':Price})
```
