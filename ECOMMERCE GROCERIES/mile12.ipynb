```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
Name=[]
Price=[]

urls=["https://mile12mart.com/product-category/fruits/","https://mile12mart.com/product-category/meat/","https://mile12mart.com/product-category/pepper/","https://mile12mart.com/product-category/tubers/","https://mile12mart.com/product-category/vegetables/","https://mile12mart.com/product-category/grains/"]
for url in urls:
    driver.get(url)
    results = driver.page_source
    soup = BeautifulSoup(results,"lxml")

    products=soup.find_all('div', class_="iq-product-title-price-holder")
    for item in products:
        name=item.h5.text
        Name.append(name)
        #print(Name)
        #price
        for p in soup.find_all('div', class_="iq-pricing-details"): 
            price=p.findAll('span')[0].text.strip()
        if price is not None:
             price= price   
        else:
             price = np.nan
        Price.append(price)
        #print(Price)


```

```python
items=pd.DataFrame({
    'Name':Name,
    'Price':Price
})
```
